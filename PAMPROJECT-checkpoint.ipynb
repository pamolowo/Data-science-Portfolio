```python
df['Self_Employed'].replace(np.nan, 'No', inplace=True)
df['Credit_History'].replace(np.nan, '1.0', inplace=True)

df['Gender']= df['Gender'].fillna(df['Gender'].mode()[0])
df['Married']= df['Married'].fillna(df['Married'].mode()[0])
df['Dependents']= df['Dependents'].fillna(df['Dependents'].mode()[0])
df.head()

```

```python
avg_LoanAmount = df["LoanAmount"].astype("float").mean(axis=0)
print(avg_LoanAmount)
```

```python
df["LoanAmount"].replace(np.nan, avg_LoanAmount, inplace = True)
df.head()
```
